```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
gp = np.array([[1,.03],[10,.0007],[20,.0002],[600,.0002],[3000,.00003],[1e5,.00003]])
gp[:,1] = gp[:,1]*1e6 #in uT
skytem = np.array([[25,.65],[75,.3],[5000,3e-3]])

toaster = np.r_[60,.38]
mixer = np.r_[60,5.25]
coffee = np.r_[60,.105]
```

```python
SMALL_SIZE = 10
MEDIUM_SIZE = 12
BIGGER_SIZE = 12

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
f,ax = plt.subplots(1,figsize=(5,4))
plt.plot(skytem[:,0],skytem[:,1],'k',label='Calculated SkyTEM level')
plt.plot(gp[:,0],gp[:,1],label='General Public reference level')
plt.plot(gp[:,0],gp[:,1]*10,'--',label='Occupational reference level')
plt.scatter(toaster[0],toaster[1],marker='.',s=200,c='k')
plt.scatter(mixer[0],mixer[1],marker='.',s=200,c='k')
plt.scatter(coffee[0],coffee[1],marker='.',s=200,c='k')


sep=100
plt.annotate('mixer',xy=mixer,xytext= mixer       + np.r_[sep,10],arrowprops={'width':1,'headwidth':5,'color':'k'})
plt.annotate('toaster',xy=toaster,xytext= toaster + np.r_[sep,.5],arrowprops={'width':1,'headwidth':5,'color':'k'})
plt.annotate('coffee machine',xy=coffee,xytext= coffee    + np.r_[sep,-.095],arrowprops={'width':1,'headwidth':5,'color':'k'})


plt.xscale('log')
plt.yscale('log')
plt.xlim(min(skytem[:,0]),max(skytem[:,0]))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnetic Flux Density ($\mu T$)')
plt.legend()
plt.savefig('/Users/ianpg/Dropbox/TempShared/GAP_files/EMF_levels.png',dpi=300,bbox_inches='tight')
```
