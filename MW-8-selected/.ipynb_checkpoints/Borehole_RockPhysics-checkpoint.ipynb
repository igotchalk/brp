```python
from pathlib import Path
import os
import sys
import numpy as np
from matplotlib import pyplot as plt
import lasio
```

```python
workdir = Path('../')
lasdir = workdir.joinpath('las_logs')

mw8_fres_fname = 'mw8_fres.las'
mw8_tmp_fname = 'mw8_tmp.las'
mw8_all_fname = 'MW-8D.las'

mw8_all  = lasio.read(lasdir.joinpath(mw8_all_fname).as_posix())
mw8_fres = lasio.read(lasdir.joinpath(mw8_fres_fname).as_posix())
mw8_tmp  = lasio.read(lasdir.joinpath(mw8_tmp_fname).as_posix())
```

```python
fres_halffoot.size - mw8_tmp['TMP'].size
```

```python
fres_dept = mw8_fres['DEPT']
mw8_dept = mw8_all['DEPT']
fres_halffoot = mw8_fres['FRES'][ mw8_fres['DEPT']%0.5==0]
fres_halffoot[fres_halffoot < 0] =.001
fres_dept_halffoot = fres_dept[ mw8_fres['DEPT']%0.5==0]

#find indicies in the mw8 log that match the fres log
fresdept2mw8dept = [np.where(ind==mw8_dept)[0] for ind in list(fres_dept_halffoot)]
fresdept2mw8dept = np.r_[[np.asscalar(ind) if len(ind)==1 else np.nan for ind in fresdept2mw8dept]]
mask = ~np.isnan(fresdept2mw8dept)


#Mask out nan values
fresdept2mw8deptma = fresdept2mw8dept[mask].astype('int32')
fres_halffootma = fres_halffoot[mask]
mw8_tmpma = mw8_tmp['TMP'][mask]
```

```python

mw8_fres_data = np.ones(len(mw8_dept))*-999.25
mw8_tmp_data = np.ones(len(mw8_dept))*-999.25

mw8_fres_data[mw8_fres_data==-999.25]  = np.nan
mw8_fres_data[fresdept2mw8deptma] = fres_halffootma

mw8_tmp_data[mw8_tmp_data==-999.25]  = np.nan
mw8_tmp_data[fresdept2mw8deptma] = mw8_tmpma

#Add the curve and export to csv
mw8_all.add_curve('FRES', mw8_fres_data, unit='OHMM')
mw8_all.add_curve('TMP', mw8_tmp_data, unit='DEGF')

mw8_all.to_csv(lasdir.joinpath('export_mw8.csv').as_posix())
```

```python
#Convert to DF and mask for plots

df = mw8_all.df()
mask = (~np.isnan(df['FRES'])  & (df.index < 340) & (df.index > 20) )
```

###  Archie's ?  
$$R_o = F*R_w $$
$$R_o = \phi^{-m}*R_w $$
### Arp's Law in degrees Fahrenheit
$$R_{w2} = R_{w1}\frac{T_1 + 6.77}{T_2 + 6.77}$$

```python
ref_temp= 77
df['FRESc'] = df['FRES']*(df['TMP']+6.77)/(ref_temp + 6.77)

plt.scatter(df['FRES'][mask],df['FRESc'][mask],label='data',c='red')
plt.plot(np.arange(1,25),np.arange(1,25),label='one-to-one')
plt.legend()
plt.xlabel('Original Fluid Res.')
plt.ylabel('Transformed Fluid Res.');
```

```python
from scipy import stats

def x_y_regression(x,y):
    slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
    plt.set_cmap('viridis_r')
    f,ax = plt.subplots(1,figsize=(7,4))
    plt.scatter(x,y,c=df.index[mask])
    plt.plot(x,slope*x + intercept,'--k')
    plt.xlabel('Fluid Res. (temp-corrected) ($\Omega-m$)')
    plt.ylabel('Bulk Res. ($\Omega-m$)')
    plt.text(x.mean(),y.mean()*2,'F = {:.2f} \n$R^2$ = {:.2f}'.format(slope,r_value**2))
    cb = plt.colorbar()
    cb.set_label('Depth (ft)', rotation=0)
    plt.title('MW-8: Fluid Resistivity vs Bulk Resistivity')

x = df['FRES'][mask]
y = df['RILD'][mask]
x_y_regression(x,y)

x = df['FRES'][mask]
y = df['RILM'][mask]
x_y_regression(x,y)
```
